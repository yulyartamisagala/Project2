Pre-processing data

1. Cleansing

```python
import re
ulasan = pd.read_csv("1000 data ulasan Tiktok.csv")

# Fungsi untuk membersihkan ulasan
def cleaningulasan(ulasan):
    ulasan = re.sub(r'@[A-Za-z0-9]+', ' ', ulasan)
    ulasan = re.sub(r'#[A-Za-z0-9]+', ' ', ulasan)
    ulasan = re.sub(r"http\S+", ' ', ulasan)
    ulasan = re.sub(r'[0-9]+', ' ', ulasan)
    ulasan = re.sub(r"[-()\"#/@;:<>{}'+=~|.!?,_]", " ", ulasan)
    ulasan = ulasan.strip()
    return ulasan

# Fungsi untuk menghapus emoji
def clearEmoji(ulasan):
    return ulasan.encode('ascii', 'ignore').decode('ascii')

# Fungsi untuk mengganti karakter yang berulang lebih dari dua kali
def replaceTOM(ulasan):
    pola = re.compile(r'(.)\1{2,}', re.DOTALL)
    return pola.sub(r'\1', ulasan)

# Fungsi untuk melakukan case folding
def casefoldingText(ulasan):
    return ulasan.lower()

# Menerapkan fungsi pembersihan pada DataFrame
ulasan['Cleaning'] = ulasan['content'].apply(cleaningulasan)
ulasan['HapusEmoji'] = ulasan['Cleaning'].apply(clearEmoji)
ulasan['3_Lebih'] = ulasan['HapusEmoji'].apply(replaceTOM)
ulasan['CaseFolding'] = ulasan['3_Lebih'].apply(casefoldingText)

print(ulasan.head())
ulasan.to_csv("prepocessing.csv", index=False)
```

2. Tokenizing

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
# Tokenizing
nltk.download('punkt')
def tokenizingText(ulasan):
    ulasan = word_tokenize(ulasan)
    return ulasan
ulasan['Tokenizing'] = ulasan['CaseFolding'].apply(tokenizingText)
print(ulasan.head())
```

3. Normalizing

```python
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/Data/kamus_normalisasi.txt'

kamusNormalisasi = {}
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip()
        if line:  # Pastikan baris tidak kosong
            parts = line.split('\t')
            if len(parts) >= 2:
                key = parts[0].strip()
                value = '\t'.join(parts[1:]).strip()
                kamusNormalisasi[key] = value
            else:
                print(f"Baris tidak sesuai format: {line}")
```

```python
# Fungsi untuk mengonversi kata-kata ke dalam kamus normalisasi
def convertToNormalisasi(tokens):
    global kamusNormalisasi
    pattern = re.compile(r'\b(' + '|'.join(re.escape(key) for key in kamusNormalisasi.keys()) + r')\b')
    content = []
    for kata in tokens:
        filterNormalisasi = pattern.sub(lambda x: kamusNormalisasi[x.group()], kata)
        content.append(filterNormalisasi.lower())
    return content
ulasan['Normalisasi'] = ulasan['Tokenizing'].apply(convertToNormalisasi)

print(ulasan.head())
```

4. Stopwords Removal

```python
nltk.download('stopwords')

# Define Indonesian stopwords and add additional stop words
stop_words = set(stopwords.words('indonesian'))
stop_words.update(["yg", "nya", "tik", "tok", "tiktok"])

# Define the function to remove stop words
def remove_stop_words(ulasan):
    # Ensure the input is a list
    if isinstance(ulasan, str):
        ulasan = eval(ulasan)
    return [word for word in ulasan if word not in stop_words]

# Apply the function to the Tokenizing column
ulasan['RemoveStopWords'] = ulasan['Normalisasi'].apply(remove_stop_words)

# Display the first few rows of the DataFrame
print(ulasan.head())
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in ulasan['RemoveStopWords']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])

def stemmingText(document):
    return [term_dict[term] for term in document]

ulasan['Stemming'] = ulasan['RemoveStopWords'].swifter.apply(stemmingText)
```

```python
# Load the preprocessed data
ulasan = pd.read_csv("Pelabelan Data.csv")

# Define the function to join words in each review
def join_words(words):
    # Ensure the input is a string representation of a list
    if isinstance(words, str):
        words = eval(words)
    return ' '.join(words)

# Apply the function to the stemming column
ulasan['joined_stemming'] = ulasan['Stemming'].apply(join_words)

# Apply the function to label sentiment
ulasan['sentiment'] = ulasan['score'].apply(label_sentiment)

# Generate wordcloud for each sentiment
sentiments = ['negatif', 'positif']

for sentiment in sentiments:
    text = ' '.join(ulasan[ulasan['sentiment'] == sentiment]['joined_stemming'])

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Wordcloud for {sentiment} sentiment')
    plt.show()
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Menggunakan data yang sudah dibersihkan dan dilabeli sentiment
corpus = ulasan['joined_stemming'].tolist()
labels = ulasan['sentiment']

# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Maksimum 1000 fitur kata

# Melakukan transformasi TF-IDF pada corpus
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Membagi data menjadi data latih (80%) dan data uji (20%)
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, labels, test_size=0.2, random_state=42)
```

Klasifikasi dengan SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Inisialisasi model SVM
svm_model = SVC(kernel='linear')

# Melatih model SVM dengan data latih
svm_model.fit(X_train, y_train)

# Memprediksi sentiment pada data uji
y_pred = svm_model.predict(X_test)

# Evaluasi hasil prediksi
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

Prediksi sentimen dari kalimat baru

```python
# Fungsi untuk memprediksi sentimen dari kalimat baru
def predict_sentiment(new_sentence):
    cleaned_sentence = cleaningulasan(new_sentence)
    removed_emoji = clearEmoji(cleaned_sentence)
    normalized = casefoldingText(removed_emoji)
    replaced_tom = replaceTOM(normalized)
    tokens = tokenizingText(replaced_tom)
    normalized_tokens = convertToNormalisasi(tokens)
    removed_stop_words = remove_stop_words(normalized_tokens)
    joined_sentence = ' '.join(removed_stop_words)

    transformed_sentence = tfidf_vectorizer.transform([joined_sentence])
    prediction = svm_model.predict(transformed_sentence)

    return prediction[0]
```

```python
# Contoh penggunaan fungsi predict_sentiment
new_sentence = "jelek banget apknya bikin memori penuh"
predicted_sentiment = predict_sentiment(new_sentence)
print(f'Sentimen untuk kalimat "{new_sentence}" adalah {predicted_sentiment}')
```
